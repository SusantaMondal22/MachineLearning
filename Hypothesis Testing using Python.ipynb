# 4)a. Do charges of people who smoke differ significantly  from the people who don't? 

```python
from scipy.stats import ttest_1samp, ttest_ind, mannwhitneyu, levene, shapiro
from statsmodels.stats.power import ttest_power
```

```python
# Seperating the data into 2 groups  Smoker  and non smoker
#group1 =df[df['smoker'] =='yes']
group1_smoker = df.where(df.smoker== 'yes').dropna()['charges']
group1_nonsmoker = df.where(df.smoker== 'no').dropna()['charges']
t_statistic, p_value = ttest_ind(group1_smoker, group1_nonsmoker)
print(t_statistic, p_value)
```

```python
# two-sample t-test
# null hypothesis: the two groups have the same mean
# p_value < 0.05 => alternative hypothesis:
# they don't have the same mean at the 5% significance level
print ("two-sample t-test p-value=", p_value)
if p_value < 0.05:
    print(f'With a p-value of {round(p_value,4)} the do not have same mean. So We reject the alternate')
else:
    print(f'With a p-value of {round(p_value,4)} the have same meant. So We fail to reject the alternate')
```

# 4)b. Does bmi of males differ significantly from that of  females? 
Ho : Variation in bmi of men and women is equal same
    
Ha : Variation in bmi of men is greater than it is in women

```python
xage = df[df['age'] == 18]
xage.sex.value_counts()
```

```python
sample_male = xage[xage['sex'] == 'male'].bmi.iloc[:-3]   #excluding the last two elements to match the size 2 samples
sample_female = xage[xage['sex'] == 'female'].bmi
print (sample_male,sample_female)
```

```python
v1, v2 = np.var(sample_female) , np.var(sample_male)
print(v1,v2)
```

```python
n = 33  # number of samples
dof = n - 1  # degrees of freedom
alpha = 0.05  # significance level
#For chi square table: https://people.smp.uq.edu.au/YoniNazarathy/stat_models_B_course_spring_07/distributions/chisqtab.pdf
chi_critical = 46.19    # critical chi_squared statistic. From the table

chi = (dof*v1)/v2
print(chi)
if chi < chi_critical:
    print("Since the test statistic is less than the critical value, we fail to reject the null")
else:
    print("Since the test statistic is more than the critical value, we reject the null")
```

# 4)c. Is the proportion of smokers significantly different  in different genders? 

Ho=propotion of smokers are equal

Ha=The to proportion are not equal

```python
f_smokers = df[df['sex'] == 'female'].smoker.value_counts()[1]  # number of female smokers
m_smokers = df[df['sex'] == 'male'].smoker.value_counts()[1] # number of male smokers
n_females = df.sex.value_counts()[1] # number of females in the data
n_males = df.sex.value_counts()[0] #number of males in the data
```

```python
print([f_smokers, m_smokers] , [n_females, n_males])
```

```python
print(f' Proportion of smokers in female, male = {round(115/662,2)}, {round(159/676,2)} respectively')
```

#Need to see if the propotion difference in significantly differe in term of statistics

```python
from statsmodels.stats.proportion import proportions_ztest

stat, pval = proportions_ztest( [f_smokers, m_smokers] , [n_females, n_males])

if pval < 0.05:
    print(f'With a p-value of {round(pval,4)} the difference is significant. So We reject the null')
else:
    print(f'With a p-value of {round(pval,4)} the difference is not significant. So We fail to reject the null')
```

# 4)d. Is the distribution of bmi across women with no  children, one child and two children, the same ? 


# Step 1: State the null and alternative hypothesis:

### Step 1: State the null and alternative hypothesis:

* $H_0$: $\mu1$ = $\mu2$ = $\mu3$
* $H_A$: At least one $\mu$ differs 

Here $\mu1$, $\mu2$ and $\mu3$ are the mean BMI of women with  0, 1 and 2 children respectively.

# Step 2: Decide the significance level

Here we select α = 0.05

# A side by side boxplot is one of the best way to compare group locations, spreads and shapes.

```python
df0= df[df['sex']=='female'].values
df1 =df[df['children']== 0]
df2 =df[df['children']== 1]
df3 =df[df['children']== 2]
measure_bmi_df = pd.DataFrame()
measure_bmi_df= measure_bmi_df.append(df1) 
measure_bmi_df = measure_bmi_df.append(df2) 
measure_bmi_df = measure_bmi_df.append(df3) 
```

```python
sns.boxplot(x = "children", y = "bmi", data = measure_bmi_df)
plt.title('BMI of women with no of Childrens')
plt.show()
```

# Step 3: Identify the test statistic

Here we have three groups. Analysis of variance can determine whether the means of three or more groups are different. ANOVA uses F-tests to statistically test the equality of means.

# Step 4: Calculate p value using ANOVA table


statsmodels.formula.api.ols creates a model from a formula and dataframe


statsmodels.api.sm.stats.anova_lm gives an Anova table for one or more fitted linear models

```python
import statsmodels.api         as     sm
from   statsmodels.formula.api import ols
 
mod = ols('bmi ~ children', data = measure_bmi_df).fit()
aov_table = sm.stats.anova_lm(mod, typ=2)
print(aov_table)
```

# Step 5: Decide to reject or accept null hypothesis

In this example, p value is 0.396458 and it is more than our chosen level of signifance at 5%
So the statistical decision is to fail to reject the null hypothesis at 5% level of significance.
So there is sufficient evidence to fail to reject the null hypothesis that bmiacross women with no children, one child and two children is different from others .
